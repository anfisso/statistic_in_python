<p style='text-align:center;font-size:18px'>Формирование гипотез</p>
<p style='font-size:17px'><br>Нулевая гипотеза - сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом<br>Альтернативная гипотеза - занятость сотрудников на рабочем месте не зависит от их оклада </p>
    

```python
# делим df на низкую и среднюю зарплаты   
salary_low = hr[hr['salary'] == 'low']['average_montly_hours']
salary_high = hr[hr['salary'] == 'high']['average_montly_hours']
```

```python
from scipy import stats 

p_value = stats.ttest_ind(salary_high, salary_low, equal_var=False)
alpha = 0.05
print(f'p-value - {round(p_value.pvalue, 5)}')

if p_value.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, не одинаковы')
else:
    print('Не отвергаем нулевую гипотезу, сотрудники работают независимо от их оклада ')
```

- Доля сотрудников с повышением за последние 5 лет


```python
sum_ = hr['left'].count()
left = hr.groupby(hr['left'] == 1)['promotion_last_5years'].sum()
share_y = left[1] / sum_ * 100
share_n = left[0] / sum_ * 100
```

- Средняя степень удовлетворенности


```python
satisfaction = hr.groupby(hr['left'] == 1)['satisfaction_level'].mean()
```

- Среднее количество проектов- Chtlytt rjkbxtcndj ghjtrnjd

```python
projects = hr.groupby(hr['left'] == 1)['number_project'].mean()
```

## уволившиеся сотрудники 

```python
print(f'Доля сотрудников с повышением за последние 5 лет - {round(share_y, 1)}%')
print(f'Средняя степень удовлетворенности {satisfaction[1].round(5)}')
print(f'Среднее количество проектов - {projects[1].round(3)}')
```

## не уволившиеся сотрудники 

```python
print(f'Доля сотрудников с повышением за последние 5 лет - {round(share_n, 1)}%')
print(f'Средняя степень удовлетворенности {satisfaction[0].round(5)}')
print(f'Среднее количество проектов - {projects[0].round(3)}')
```

```python
hr_left = hr.drop(['department', 'salary', 'left'], axis=1)

hr_left.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(hr_left, hr['left'], test_size=0.25)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
y_pred = lda.predict(X_test)
y_pred
```

```python
# качество модели 
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
